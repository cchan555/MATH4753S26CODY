```python
# Cody Chan
# Lab 2 Assignment: Text Pre-Processing using spaCy

# Load in required libraries and import them
from requests.models import Response
!pip install spacy
!python -m spacy download en_core_web_sm
import spacy
import requests # Library to get data from internet

# Load in and read file from GitHub
url = "https://raw.githubusercontent.com/cchan555/lis4693/refs/heads/main/lab-1/transcript"
response = requests.get(url) # Sends HTTP request to retrieve file
response.raise_for_status() # Raise exception for HTTP errors
text = response.text # Extracts text content

print("Number of Characters:", len(text)) # Print the number of characters
print(text[:100]) # Print the first 100 characters

# Sentence Segmentation
nlp = spacy.blank("en") # Create blank English pipeline
nlp.add_pipe("sentencizer") # Add rule for boundary
doc = nlp(text) # Process text
sentences = list(doc.sents) # Convert sentence into list
print("Number of sentences:", len(sentences))
print("\nFirst 3 sentences:")
for sent in sentences[:3]: # Loop through and print up the the # specified
  print(sent)

# Word count and Token Analysis
# Create list of alphabetic tokens and convert to lowercase
words = [token.text.lower() for token in doc if token.is_alpha]
print("Total Words:", len(words))
print("Unique Words:", len(set(words)))

# Most Frequent Words
from collections import Counter # Import Counter library
word_freq = Counter(words) # Counts for each word
print("Top 10 most frequent words:")
for word, count in word_freq.most_common(10): # Loop through and print
  print(word, count)

# Use Full spaCy Pipeline (Entity)
nlp2 = spacy.load("en_core_web_sm") # Load fill English model
doc2 = nlp2(text) # Processes text
print("Named Entities:")
for ent in doc2.ents: # Loop through and print
  print(ent.text, ent.label_)
print("Total number of Entities:", len(doc2.ents))

# PhraseMatcher
from spacy.matcher import PhraseMatcher # Load in library for phraseMatcher
matcher = PhraseMatcher(nlp2.vocab, attr = "LOWER") # Case sensitive search
phrases = ["snowball", "county", "oklahoma"] # Three phrases related to transcript
patterns = [nlp2(p) for p in phrases] # Convert phrases into spaCy
matcher.add("Terms", patterns) # Add patterns for label "Terms"
matches = matcher(doc2)
print("Matches found:")
for match_id, start, end in matches: # Loop through and print
  print(doc2[start:end])
  print("Sentence:", doc2[start].sent)
```
